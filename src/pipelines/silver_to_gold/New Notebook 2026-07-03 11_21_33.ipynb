```sql
%python
import os

# Catalog & Schema
catalog = "retail_project"
schema = "retail_gold"

# Tables to export
tables = [
    "calendar",
    "dim_customer",
    "dim_product",
    "fact_inventory",
    "fact_sales"
]

# Output folder
output_dir = "/Workspace/Repos/github/retail-data-platform-databricks/src/pipelines/silver_to_gold"

os.makedirs(output_dir, exist_ok=True)

for table in tables:
    df = spark.table(f"{catalog}.{schema}.{table}")

    output_file = os.path.join(output_dir, f"{table}.csv")

    # Export as CSV
    df.toPandas().to_csv(output_file, index=False)

    print(f"✅ Exported: {output_file}")

print("🎉 All tables exported successfully!")
```
